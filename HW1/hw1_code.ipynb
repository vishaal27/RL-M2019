# Q1. Nonstationary Environment: Sample Average action-value method

```python
import numpy as np
import random
import matplotlib.pyplot as plt

n_runs=500
epsilon=0.1
n_steps=10000

each_step_rewards=[0]*n_steps
optimal_chosen_percentage_for_plot=[0]*n_steps

for run_no in range(n_runs):
    if(run_no%100==0):
        print('Run number: '+str(run_no))
    n_bandits=10
    q_values=[0]*n_bandits
    n_values=[0]*n_bandits
    
    q_start_values=[]
    norm=np.random.normal(0, 1)
    for i in range(n_bandits):
        q_start_values.append(norm)
        
    
    n_optimal_chosen=0
    
    for i in range(n_steps):
        
        q_start_values+=np.random.normal(0, 0.01, n_bandits)

        prob=random.random()
        if(prob<epsilon):
            action_chosen=random.randint(0, n_bandits-1)
        else:
            action_chosen=np.argmax(q_values)
        
        optimal_action=np.argmax(q_start_values)
        if(optimal_action==action_chosen):
            n_optimal_chosen+=1
            optimal_chosen_percentage_for_plot[i]+=1
            
        reward=np.random.normal(q_start_values[action_chosen], 1)  #changed
        each_step_rewards[i]+=reward
        
        n_values[action_chosen]+=1
        q_values[action_chosen]+=(1/n_values[action_chosen])*(reward-q_values[action_chosen])

plot_rewards_sa=list(np.asarray(each_step_rewards)/n_runs)
plot_optimal_percentage_sa=list(np.asarray(optimal_chosen_percentage_for_plot)/n_runs)
```

```python
plt.figure(1)
plt.plot(plot_rewards_sa)
plt.xlabel('steps')
plt.ylabel('average reward')
plt.show()

plt.figure(2)
plt.plot(plot_optimal_percentage_sa)
plt.xlabel('steps')
plt.ylabel('% optimal action')
plt.show()
```
